## Adding pickup and dropoff neighborhoods

Working with longitude and latitude values is not very practical. They do not vary much. For example, std for pickup latitude = 0.032565. </br>
One idea is calculating the distance between pickup and dropoff locations. </br>
Another idea is adding the neighborhoods to which pickup and dropoff locations belong using the geopandas library. </br>
However, executing the code to do this takes some time. That is why I execute it once here in this notebook and save the resulting dataframe as a .csv file to be used later.

I downloaded the borough geometrical boundaries data from https://data.cityofnewyork.us/City-Government/Borough-Boundaries/tqmj-j8zm

```python
nyc_bound = gpd.read_file('Borough Boundaries.geojson')
```

```python
nyc_bound.head()
```

```python
nyc_bound["boro_name"].unique()
```

```python
# Testing one example: the pickup location of the first row
longitude, latitude = -73.953918, 40.778873 
point = Point(longitude, latitude)
for idx, row in nyc_bound.iterrows():
        if row['geometry'].contains(point):
            boro = row['boro_name']
            print(boro)
```

```python
def get_boro(longitude, latitude):
    point = Point(longitude, latitude)
    boro = ''
    for idx, row in nyc_bound.iterrows():
        if row['geometry'].contains(point):
            boro = row['boro_name']
    return boro
```

```python
def assign_boro_pickup(row):
    return get_boro(row['pickup_longitude'], row['pickup_latitude'])

def assign_boro_dropoff(row):
    return get_boro(row['dropoff_longitude'], row['dropoff_latitude'])    
```

I used the joblib library to parallelize the computation.

```python
num_cores = -1  # Use all available CPU cores
pickup_boros = Parallel(n_jobs=num_cores)(delayed(assign_boro_pickup)(row) for _, row in df.iterrows())
```

```python
df['pickup_boro'] = pickup_boros
```

```python
df["pickup_boro"].value_counts()
```

Note that the empty pickup_boro value means that the location is not found in the boros boundary file. They must be locations outside NYC.

```python
df[df["pickup_boro"]==""].shape
```

```python
dropoff_boros = Parallel(n_jobs=num_cores)(delayed(assign_boro_dropoff)(row) for _, row in df.iterrows())
```

```python
df['dropoff_boro'] = dropoff_boros
```

```python
df['dropoff_boro'].value_counts()
```

```python
df.to_csv("nyc_taxi_trip_duration_with_neighborhood.csv", index=False)
```

## Playing with geographical data with geopandas
